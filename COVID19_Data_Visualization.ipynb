## 📌 Project Description

This project focuses on analyzing and visualizing global COVID-19 data using Python and data visualization libraries. The goal of the project is to explore patterns and trends in confirmed cases, deaths, recoveries, and active cases across different countries and time periods.

Using various visualization techniques such as line charts, bar charts, pie charts, scatter plots, histograms, and heatmaps, the project presents insights into how the pandemic evolved globally and which regions were most affected.

The visualizations help transform complex datasets into clear and meaningful insights that can support better understanding of pandemic trends and global health impacts.

## 🎯 Project Objectives

* Analyze global COVID-19 case trends over time
* Identify the most affected countries based on confirmed cases
* Compare relationships between confirmed cases and deaths
* Explore distribution of active cases across countries
* Visualize correlations between major COVID-19 variables
* Present insights using effective data visualization techniques


```python
#Import Visualization Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
#Load Dataset
df = pd.read_csv("covid_19_clean_complete.csv")
df.head()
```

```python
#Data Cleaning
df['Date'] = pd.to_datetime(df['Date'])
df['Province/State'] = df['Province/State'].fillna('Unknown')
df.isnull().sum()
```

```python
#Data Preparation
# Global cases over time
global_cases = df.groupby('Date')[['Confirmed','Deaths','Recovered','Active']].sum()

# Country wise summary
country_cases = df.groupby('Country/Region')[['Confirmed','Deaths','Recovered','Active']].max()
```

```python
#Global COVID-19 Trend (Line Chart)
plt.figure(figsize=(12,6))

plt.plot(global_cases.index, global_cases['Confirmed'], label='Confirmed Cases')
plt.plot(global_cases.index, global_cases['Deaths'], label='Deaths')
plt.plot(global_cases.index, global_cases['Recovered'], label='Recovered')

plt.title("Global COVID-19 Cases Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Cases")
plt.legend()

plt.show()
```

```python
#Top 10 Countries with Highest Confirmed Cases (Horizontal Bar Chart)
top_cases = country_cases.sort_values(by='Confirmed', ascending=False).head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top_cases['Confirmed'],
    y=top_cases.index,
)

plt.title("Top 10 Countries with Highest Confirmed COVID-19 Cases")
plt.xlabel("Total Confirmed Cases")
plt.ylabel("Country")

plt.show()
```

```python
#Global Case Distribution (Pie Chart)
labels = ['Confirmed','Deaths','Recovered']

values = [
    global_cases['Confirmed'].max(),
    global_cases['Deaths'].max(),
    global_cases['Recovered'].max()
]

plt.figure(figsize=(7,7))

plt.pie(values,
        labels=labels,
        autopct='%1.1f%%',
        startangle=90)

plt.title("Global COVID-19 Case Distribution")

plt.show()
```

```python
#Confirmed vs Deaths Relationship (Scatter Plot)
plt.figure(figsize=(10,6))

plt.scatter(country_cases['Confirmed'],
            country_cases['Deaths'],
            alpha=0.6)

plt.title("Confirmed Cases vs Deaths by Country")
plt.xlabel("Confirmed Cases")
plt.ylabel("Deaths")

plt.show()
```

```python
#Active Cases Distribution (Histogram)
plt.figure(figsize=(10,6))

plt.hist(country_cases['Active'], bins=30)

plt.title("Distribution of Active COVID-19 Cases")
plt.xlabel("Active Cases")
plt.ylabel("Number of Countries")

plt.show()
```

```python
#Correlation Between Variables (Heatmap)
plt.figure(figsize=(8,6))

sns.heatmap(
    country_cases[['Confirmed','Deaths','Recovered','Active']].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Between COVID-19 Variables")

plt.show()
```

```python
#India vs USA Cases Comparison (Multi-Line Chart)
countries = ['India','US']

compare = df[df['Country/Region'].isin(countries)]

compare_cases = compare.groupby(['Date','Country/Region'])['Confirmed'].sum().unstack()

compare_cases.plot(figsize=(12,6))

plt.title("COVID-19 Confirmed Cases: India vs USA")
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")

plt.show()
```
